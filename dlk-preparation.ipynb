# Deutscher Lyrik Korpus

```python
import os
import json
from tqdm.notebook import tqdm, trange
import seaborn as sns
import pandas as pd
import pickle as pkl
```

```python
path = 'F:/$Daten/datasets/DLK-master/DLK/standard/'
with open(path + 'dlk.v5.german.poetry.corpus.full.json', 'rb') as f:
    d = json.load(f)
```

```python
list(d.keys())
d['dta.poem.1']#['metadata']['pub_year']
```

```python
df = pd.DataFrame(columns=['Pub year', 'Author', 'Title', 'Text', 'Encoded'])
for k in tqdm(d.keys()):
    metadata = d[k]['metadata']
    pub_year = metadata['pub_year']
    author = metadata['author']['name']
    title = metadata['title']

    poem = d[k]['poem']
    text = ''
    for stanza in poem:
        for k, v in poem[stanza].items():
            if 'line' in k:
                text = text + v['text'] + '\n'
        text = text + '\n'

    df.loc[len(df.index)] = [pub_year, author, title, text, '']
```

```python
path = 'F:/$Daten/datasets/DLK-master/'
df.to_csv(path+'dlk.csv', index=False)
```

### Bytepair Encoding

```python
import bpe
```

```python
text = []
for i in trange(0, len(df), 100):
    text.append(df['Text'][i])
text = '\n'.join(text)
```

```python
bpc = bpe.create_code(text, 10000, nmin=3)
```

```python
bpc
```

### Dataframe erstellen und speichern

```python
df = pd.DataFrame(columns=['Pub year', 'Author', 'Title', 'Text', 'Encoded'])
for k in tqdm(d.keys()):
    metadata = d[k]['metadata']
    pub_year = metadata['pub_year']
    author = metadata['author']['name']
    title = metadata['title']

    poem = d[k]['poem']
    text = ''
    for stanza in poem:
        for k, v in poem[stanza].items():
            if 'line' in k:
                text = text + v['text'] + '\n'
        text = text + '\n'

    encoded = bpe.encode(text, bpc)

    df.loc[len(df.index)] = [pub_year, author, title, text, encoded]
```

```python
path = 'F:/$Daten/datasets/DLK-master/'
df.to_csv(path+'dlk.csv', index=False)
```

```python
N = len(df)
train_inds = [ i for i in range(N) if i%10!=0 ]
val_inds = [ i for i in range(N) if i%10==0 ]
train_df = df.iloc[train_inds]
val_df = df.iloc[val_inds]
```

```python
path = 'F:/$Daten/datasets/DLK-master/'
train_df.to_csv(path+'dlk_train.csv', index=False)
val_df.to_csv(path+'dlk_val.csv', index=False)
```

```python
alphabet = set()
for i in trange(N):
    alphabet |= set(df['Encoded'][i])
alphabet
```
